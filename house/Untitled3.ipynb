```python
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import GradientBoostingClassifier

model1 = RandomForestClassifier(n_estimators=100, random_state=42)
model2 = GradientBoostingClassifier(n_estimators=100, random_state=42)

ensemble_model = VotingClassifier(estimators=[('rf', model1), ('gb', model2)], voting='soft')
ensemble_model.fit(X_train, y_train)
ensemble_score = ensemble_model.score(X_test, y_test)
print("Ensemble model score:", ensemble_score)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

grid_search = GridSearchCV(model1, param_grid, cv=1)
grid_search.fit(X_train, y_train)
best_params = grid_search.best_params_

print("Best parameters for Random Forest:", best_params)
```

```python
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

model3 = SVC(probability=True, random_state=42)
model4 = KNeighborsClassifier()

ensemble_model = VotingClassifier(estimators=[('rf', model1), ('gb', model2), ('svc', model3), ('knn', model4)], voting='soft')
ensemble_model.fit(X_train, y_train)
ensemble_score = ensemble_model.score(X_test, y_test)
print("Enhanced ensemble model score:", ensemble_score)
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression

base_models = [
    ('rf', RandomForestClassifier(n_estimators=100, random_state=42)),
    ('gb', GradientBoostingClassifier(n_estimators=100, random_state=42)),
    ('svc', SVC(probability=True, random_state=42)),
    ('knn', KNeighborsClassifier())
]

meta_model = LogisticRegression()
stacking_model = StackingClassifier(estimators=base_models, final_estimator=meta_model)

stacking_model.fit(X_train, y_train)
stacking_score = stacking_model.score(X_test, y_test)
print("Stacking model score:", stacking_score)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

model_pca = RandomForestClassifier(n_estimators=100, random_state=42)
model_pca.fit(X_train_pca, y_train)
pca_score = model_pca.score(X_test_pca, y_test)
print("PCA model score:", pca_score)
```

```python
from skopt import BayesSearchCV

# Define the parameter space
param_space = {
    'n_estimators': (10, 500),
    'max_depth': (1, 50),
    'min_samples_split': (2, 20),
}

# Initialize and perform Bayesian Optimization
bayes_search = BayesSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    search_spaces=param_space,
    n_iter=32,
    cv=1,
    random_state=42
)
bayes_search.fit(X_train, y_train)
best_params = bayes_search.best_params_
print("Best parameters found by Bayesian Optimization:", best_params)
```
